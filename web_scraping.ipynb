```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Base URL per la paginazione
base_url = 'https://books.toscrape.com/catalogue/page-{}.html'

# Lista per contenere tutti i dati
all_books = []

# Scorri le 50 pagine
for page in range(1, 51):
    print(f'ðŸ”„ Estrazione da pagina {page}...')
    url = base_url.format(page)
    response = requests.get(url)

    # Se la pagina non esiste o dÃ  errore, esci
    if response.status_code != 200:
        print(f' Pagina {page} non trovata (errore {response.status_code}). Interrotto.')
        break

    soup = BeautifulSoup(response.text, 'html.parser')
    books = soup.find_all('article', class_='product_pod')

    # Estrai dati per ogni libro
    for book in books:
        title = book.h3.a['title']
        price = book.find('p', class_='price_color').text
        availability = book.find('p', class_='instock availability').text.strip()
        all_books.append([title, price, availability])

# Salva i dati in un CSV
df = pd.DataFrame(all_books, columns=['Titolo', 'Prezzo', 'DisponibilitÃ '])
df.to_csv('books_all.csv', index=False)

# Mostra anteprima
print(" Estrazione completata.")
df.head()
```
