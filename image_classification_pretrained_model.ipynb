# 1. Imports
Imports Pandas to handle tabular data, loads the "id" and "gender" columns from the dataset (styles.csv), and displays the first few rows.
Sets the device to GPU (CUDA) if available

```python
import pandas as pd # Used to read/create dataframes (csv) and process tabular data

data_df = pd.read_csv("/content/fashion-product-images-small/styles.csv", usecols=["id", "gender"])
print(data_df.head())
```

```python
import torch # Main PyTorch Library
from torch import nn # Used for creating the layers and loss function
from torch.optim import Adam # Adam Optimizer
import torchvision.transforms as transforms # Transform function used to modify and preprocess all the images
from torchvision import models # import pretrained models in PyTorch library
from torch.utils.data import Dataset, DataLoader # Dataset class and DataLoader for creating the objects
from sklearn.preprocessing import LabelEncoder # Label Encoder to encode the classes from strings to numbers
import matplotlib.pyplot as plt # Used for visualizing the images and plotting the training progress
from PIL import Image # Used to read the images from the directory
import numpy as np # preprocessing and numerical/mathematical operations
import os # Used to read the images path from the directory

device = "cuda" if torch.cuda.is_available() else "cpu" # detect the GPU if any, if not use CPU, change cuda to mps if you have a mac
```

# 3. Data Manipulation and Reading Data Paths

Merges gender categories by replacing "Boys" with "Men" and "Girls" with "Women"and then prints the unique gender classes and their distribution in the dataset.

```python
data_df["gender"] = data_df["gender"].replace({
    "Boys": "Men",
    "Girls": "Women"
})

print("Classes are: ")
print(data_df["gender"].unique())
print()
print("Classes ditrubution are: ")
print(data_df["gender"].value_counts())
```

Modify the "id" column to create the full image file paths by appending the directory path and .jpg extension. Then display the first few rows of the updated dataset.

```python
data_df["id"] = "/content/fashion-product-images-small/images/" + data_df["id"].astype(str) + ".jpg"

# Filter out rows where the image file does not exist
data_df = data_df[data_df["id"].apply(os.path.exists)]

print(data_df.head())
```

Calculate the sample size for each gender category by taking a percentage of the total count if the dataset is too large.

```python
MY_SAMPLE_SIZE = 0.1

# Calculate sample size per category (10% of each gender)
sample_size = data_df["gender"].value_counts() * MY_SAMPLE_SIZE  # 10% of each group
sample_size = sample_size.astype(int)  # Convert to integer count

print(sample_size)
```

Perform the sampling

```python
# Perform stratified sampling
sampled_df = data_df.groupby("gender").apply(lambda x: x.sample(sample_size[x.name], random_state=42))

# Remove extra index added by groupby
sampled_df.reset_index(drop=True, inplace=True)

print(sampled_df["gender"].value_counts())
```

# 5. Data Split
* Training: 70%
* Testing: 30%

```python
train=sampled_df.sample(frac=0.7,random_state=7) # Create training of 70% of the data
test=sampled_df.drop(train.index) # Create testing by removing the 70% of the train data which will result in 30%
```

# 6. Preprocessing Objects
First, we use LabelEncoder to convert string class labels into numeric values (e.g., "Men", "Women" â†’ 0, 1). Then, we define a transform object to resize images, convert them to tensors, and ensure they are of type float for consistent preprocessing.

```python
label_encoder = LabelEncoder() # Encode the string classes to numeric

transform = transforms.Compose([
    transforms.Resize((128, 128)), # One size for all images
    transforms.ToTensor(), # Convert images to PyTorch tensors
    transforms.ConvertImageDtype(torch.float), # The values are in floating point numbers
    ]) # Transform all images into one clear format (preprocess all images to same properties)
```

# 7. Custom Dataset Class
Define a custom dataset class CustomImageDataset that inherits from Dataset. It loads images and their labels, applies transformations, and returns the preprocessed image and corresponding label encoded into numeric.

```python
class CustomImageDataset(Dataset):
  def __init__(self, dataframe, transform=None):
    self.dataframe = dataframe
    self.transform = transform
    self.labels = torch.tensor(label_encoder.fit_transform(dataframe['gender'])).to(device)

  def __len__(self):
    return self.dataframe.shape[0]

  def __getitem__(self, idx):
    img_path = self.dataframe.iloc[idx, 0]
    label = self.labels[idx]
    image = Image.open(img_path).convert('RGB')
    if self.transform:
      image = (self.transform(image)/255).to(device)

    return image, label
```

# 8. Create Dataset Objects
Create instances of CustomImageDataset for both training and testing datasets, applying the defined image transformations to preprocess the data.

```python
train_dataset = CustomImageDataset(dataframe=train, transform=transform)
test_dataset = CustomImageDataset(dataframe=test, transform=transform)
```

# 10. Hyperparameters
Setting the training hyperparameters, we defined some variables which are the batch size, number of training epochs and learning rate.

```python
LR = 1e-3
BATCH_SIZE = 4
EPOCHS = 15
```

Create data loaders for the training and testing datasets

```python
train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=True)
```

Load the pre-trained GoogleNet model with default weights

```python
googlenet_model = models.googlenet(weights='DEFAULT')
for param in googlenet_model.parameters():
  param.requires_grad = True
```

Access the fully connected (fc) layer of the pre-trained GoogleNet model. This is typically the classification layer that can be modified for custom tasks.

```python
googlenet_model.fc
```

Modify the fully connected (fc) layer of the GoogleNet model to match the number of classes in the dataset (num_classes), and move the model to the specified device (GPU or CPU).

```python
num_classes = len(data_df["gender"].unique())
googlenet_model.fc = torch.nn.Linear(googlenet_model.fc.in_features, num_classes)
googlenet_model.to(device)
```

Define the loss function as CrossEntropyLoss for multi-class classification and the optimizer as Adam with the model's parameters and specified learning rate (LR).

```python
# Define loss function and optimizer
criterion = nn.CrossEntropyLoss() # Cross Entropy Loss
optimizer = Adam(googlenet_model.parameters(), lr=LR) # Adam optimizer
```

Training

```python
from tqdm import tqdm
import time

total_loss_train_plot = []
total_acc_train_plot = []

EPOCHS = 10  # Change this based on your setup

for epoch in range(EPOCHS):
    total_acc_train = 0
    total_loss_train = 0

    # Wrap train_loader with tqdm for a progress bar
    loop = tqdm(train_loader, leave=True, desc=f"Epoch {epoch+1}/{EPOCHS}")

    start_time = time.time()  # Track time

    for inputs, labels in loop:
        optimizer.zero_grad()
        outputs = googlenet_model(inputs)
        train_loss = criterion(outputs, labels)
        total_loss_train += train_loss.item()
        train_loss.backward()

        train_acc = (torch.argmax(outputs, axis=1) == labels).sum().item()
        total_acc_train += train_acc
        optimizer.step()

        # Update tqdm progress bar
        loop.set_postfix(loss=train_loss.item(), accuracy=train_acc / len(labels))

    # Calculate and print epoch time
    epoch_time = time.time() - start_time
    print(f"Epoch {epoch+1} completed in {epoch_time:.2f} seconds")
    print()
    total_loss_train_plot.append(round(total_loss_train/1000, 4))
    total_acc_train_plot.append(round(total_acc_train/(train_dataset.__len__())*100, 4))
    print(f'Epoch {epoch+1}/{EPOCHS}, Train Loss: {round(total_loss_train/100, 4)} Train Accuracy {round((total_acc_train)/train_dataset.__len__() * 100, 4)}%')
    print()
```

```python
with torch.no_grad():
  total_loss_test = 0
  total_acc_test = 0
  for indx, (input, labels) in enumerate(test_loader):

    prediction = googlenet_model(input)

    acc = (torch.argmax(prediction, axis = 1) == labels).sum().item()
    total_acc_test += acc

print(f"Accuracy Score is: {round((total_acc_test/test_dataset.__len__())*100, 2)}%")
```
